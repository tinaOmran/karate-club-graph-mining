```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
from networkx.algorithms.link_prediction import adamic_adar_index
from node2vec import Node2Vec
from sklearn.manifold import TSNE
from sklearn.metrics import precision_recall_curve, auc, accuracy_score
from sklearn.metrics.cluster import homogeneity_score, completeness_score, v_measure_score
from sklearn.metrics import homogeneity_score, roc_auc_score
```

```python
karate_club_graph = nx.karate_club_graph()

# Graph properties
print("Number of nodes:", karate_club_graph.number_of_nodes())
print("Number of edges:", karate_club_graph.number_of_edges())
print("Average degree of nodes:", sum(dict(karate_club_graph.degree()).values()) / karate_club_graph.number_of_nodes())
print("Average clustering coefficient:", nx.average_clustering(karate_club_graph))


node_colors = [
    'red' if karate_club_graph.nodes[node]['club'] == 'Mr. Hi' else 'blue'
    for node in karate_club_graph.nodes
]

```

```python
plt.figure(figsize=(8, 8))
nx.draw(
    karate_club_graph,
    with_labels=True,
    node_color=node_colors,
    node_size=500,
    font_size=10
)
plt.title("Zachary's Karate Club Graph")
plt.show()
```

```python
#Calculate the distribution of node degrees
degrees = [degree for _, degree in karate_club_graph.degree()]
plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=range(min(degrees), max(degrees) + 1), alpha=0.7, color='blue', edgecolor='black')
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.show()
```

```python
node2vec = Node2Vec(karate_club_graph, dimensions=64, walk_length=30, num_walks=200, workers=4)
model = node2vec.fit(window=10, min_count=1, batch_words=4)
node_embeddings = {int(node): model.wv[str(node)] for node in karate_club_graph.nodes}

embedding_list = list(node_embeddings.values())

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(embedding_list)

cluster_labels = {node: cluster for node, cluster in zip(node_embeddings.keys(), clusters)}

embedding_array = np.array(embedding_list)  


tsne = TSNE(n_components=2, random_state=0)
reduced_embeddings = tsne.fit_transform(embedding_array)


plt.figure(figsize=(8, 8))
for cluster in set(clusters):  
    indices = [i for i, c in enumerate(clusters) if c == cluster]
    plt.scatter(reduced_embeddings[indices, 0], reduced_embeddings[indices, 1], label=f'Cluster {cluster}')

plt.legend()
plt.title("t-SNE Visualization of Node Embeddings")
plt.show()
```

```python
def calculate_purity(true_labels, predicted_clusters):
    
    contingency_matrix = confusion_matrix(true_labels, predicted_clusters)
    
    max_values = np.sum(np.amax(contingency_matrix, axis=0))
    
    purity = max_values / np.sum(contingency_matrix)
    return purity

true_labels = [karate_club_graph.nodes[node]["club"] for node in karate_club_graph.nodes]
true_labels = [0 if label == "Mr. Hi" else 1 for label in true_labels]  


predicted_clusters = kmeans.labels_  


purity = calculate_purity(true_labels, predicted_clusters)
print(f"Clustering Purity: {purity:.2f}")
```

```python
edges_to_remove = random.sample(list(karate_club_graph.edges()), int(0.2 * karate_club_graph.number_of_edges()))

training_graph = karate_club_graph.copy()
training_graph.remove_edges_from(edges_to_remove)

print(f"removed edges : {num_edges_to_remove} train edges :{training_graph.number_of_edges()}")
predictions = list(adamic_adar_index(training_graph))
sorted_predictions = sorted(predictions, key=lambda x: x[2], reverse=True)


pos = nx.spring_layout(karate_club_graph)  

plt.figure(figsize=(10, 10))

correct_edges = [edge for edge in edges_to_remove if edge in [(u, v) for u, v, score in sorted_predictions[:len(edges_to_remove)]]]
incorrect_edges = [edge for edge in edges_to_remove if edge not in correct_edges]


nx.draw_networkx_edges(training_graph, pos, alpha=0.3, edge_color="gray", label="Training Edges")
nx.draw_networkx_edges(karate_club_graph, pos, edgelist=correct_edges, edge_color="green", label="Correct Predictions")
nx.draw_networkx_edges(karate_club_graph, pos, edgelist=incorrect_edges, edge_color="red", label="Incorrect Predictions")

node_colors = []
for node, data in karate_club_graph.nodes(data=True):
    if data["club"] == "Mr. Hi":
        node_colors.append("blue")
    else:  
        node_colors.append("orange")
nx.draw_networkx_nodes(karate_club_graph, pos, node_color=node_colors, label="Nodes")
nx.draw_networkx_labels(karate_club_graph, pos)

plt.legend()
plt.title("Visualization of Edge Predictions")
plt.show()

y_true = [1 if edge in edges_to_remove else 0 for edge in [(u, v) for u, v, score in sorted_predictions]]
y_scores = [score for u, v, score in sorted_predictions]


precision, recall, _ = precision_recall_curve(y_true, y_scores)
pr_auc = auc(recall, precision)
auc_score = roc_auc_score(y_true, y_scores)


plt.figure(figsize=(5, 5))
plt.plot(recall, precision, label=f"Precision-Recall AUC = {pr_auc:.2f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
print(f"AUC: {auc_score:.2f}")
```
